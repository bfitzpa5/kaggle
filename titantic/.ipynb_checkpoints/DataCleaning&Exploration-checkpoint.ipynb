**Function to emulate sns pairplot single plot**

```python
def scatter(df, targetName, columnNameX, columnNameY):
    positive = df[df[targetName]==1].dropna()
    negative = df[df[targetName]==0].dropna()
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(positive[columnNameX],positive[columnNameY],color='green',marker='o',label=targetName)
    ax.scatter(negative[columnNameX],negative[columnNameY],color='red',marker='x',label='Not '+targetName)

    plt.legend()
    plt.show()
```

```python
scatter(df, "Survived","Age","Fare")
```

# Data Cleaning

### Dummy Variable for Sex

```python
df['Female'] = df.Sex.apply(lambda x: 1 if x=='female' else 0)
df[['Sex','Female']]
```

### Dummy Variable for Pclass

```python
df['FirstClass'] = df.Pclass.apply(lambda x: 1 if x==1 else 0)
df['SecondClass'] = df.Pclass.apply(lambda x: 1 if x==2 else 0)
df[['Pclass','FirstClass','SecondClass']]
```

### Dummy Variable for Port of Embarkation

```python
df['Embarked_Cherbourg'] = df.Embarked.apply(lambda x: 1 if x=='C' else 0)
df['Embarked_Queenstown'] = df.Embarked.apply(lambda x: 1 if x=='Q' else 0)
df[['Embarked','Embarked_Cherbourg','Embarked_Queenstown']]
```

### Remove Dummy Variables Previous Columns

```python
df.drop(['Embarked','Sex','Pclass'],axis=1,inplace=True)
```

### Remove Cabin 

Since it only contains 204 non-null entries, while the rest contains 714-894 non-null entries

```python
df.drop('Cabin',axis=1,inplace=True)
```

### Remove Rows with Null Values

This will remove 177 columns, since age contains 714 non-null entries, while every other column contains 891 entries

```python
df.dropna(inplace=True)
```

### Set Index as PassengerId

```python
df.set_index('PassengerId')
```
